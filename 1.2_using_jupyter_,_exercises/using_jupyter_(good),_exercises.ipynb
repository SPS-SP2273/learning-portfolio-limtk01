## 3 Timing Stuff

```python
squares = [i**2 for i in range (10)]
print (squares)
```

```python
squares = list()
for i in range (10):
    squares.append(i**2)
print (squares)
```

- Both codes give the same results

```python
%%timeit -n 100000 -r 7
squares = [i**2 for i in range (10)]
```

```python
%%timeit -n 100000 -r 7
for i in range (10):
    squares.append(i**2)
```

- to find out how much time does it take for the cell to execute the command, the system took a total of 7 runs, each run consisting of 100000 loops (ie. run the cell 100000 times, and repeat it for a total of 7 times)
- 223 ns indicates refers to the mean time to run the cell is 223 nano-seconds
- 103 ns indicates refers to the standard deviation of the spread of the timings collected
